```python
from glob import glob
# 이미지들의 주소 리스트로 만들어줌
train_img_list = glob('/content/landmark-3/train/images/*.jpg')
valid_img_list = glob('/content/landmark-3/valid/images/*.jpg')
# 리스트를 txt 파일로 생성
with open('/content/landmark-3/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')
with open('/content/landmark-3/valid.txt', 'w') as f:
  f.write('\n'.join(valid_img_list) + '\n')
```

```python
import yaml
with open('/content/landmark-3/data.yaml', 'r') as f:
  data = yaml.safe_load(f)
print(data)
data['train'] = '/content/landmark-3/train.txt'
data['val'] = '/content/landmark-3/valid.txt'
with open('/content/landmark-3/data.yaml', 'w') as f:
  yaml.dump(data, f)
print(data)
```
